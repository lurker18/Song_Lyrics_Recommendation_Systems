# KPop-KR PWI Evaluation

```python
from nltk import ngrams, FreqDist
freq_dist = FreqDist()
for doc in kpop_kr:
    doc = doc.split()
    for word in doc:
        freq_dist[word.lower()] += 1
freq_dist
```

```python
count = 0
for i in tqdm(range(4577)):
    a =len(kpop_kr[i])
    count = count + a
count
```

```python
kpop_kr_topic_list = []

for i in range(28):
    temp = []
    for j in range(10):
        temp.append(kpop_kr_topic_model.get_topic(i)[j][0])
    kpop_kr_topic_list.append(temp)

kpop_kr_topic_list
```

```python
from numpy import inf

total_pwi=[]
N = 4577
#pattern = r'\w+'
#Store total frequency of all terms in document set
F = count

k = 28 # 28 / 35 / 50 

#Probability of dj is all 1/all documents
P_dj = (1 / N)
#print(f'P_dj {P_dj}: 1 / N {N}')
#Iterate over all documents

for i in tqdm(range(k)):
    dj_wi_pwi_list = []
    for d in kpop_kr:
        words_in_t = []
        #Iterate over n topic words
        fake_topic_words = kpop_kr_topic_list[i]
        for word in fake_topic_words:
            #print('word:==========', word)
            #print(d)
            #print(word)
            tokenized_doc = [x.lower() for x in list(d.split(' '))] #Should this be a set?
            #print(f'TOKENIZED_DOC:{tokenized_doc}\n')
    
            f_ij = sum([1 for x in tokenized_doc if word == x])
            #print(f_ij)
            P_wi_dj = (f_ij / F)
            #print(f'P_wi_dj:{P_wi_dj} = (f_ij {f_ij} / F {F})\n')
            
            P_wi = freq_dist.freq(word) #전체문서에서 해당 단어가 등장할 확률
            #print(P_dj)
            try:
                if np.log(P_wi_dj/(P_wi * P_dj)) == -inf:
                    pmi = 0
                else :
                    pmi = np.log(P_wi_dj/(P_wi * P_dj))
            except:
                pass
            #print(pmi)
            
            #print(f'pmi:{pmi} = np.log(P_wi_dj {P_wi_dj}/(p_wi {p_wi} * P_dj {P_dj})\n')
            
            #pmi = np.log(p_wd) - np.log(p_w) - np.log(p_d)
    
            # p(d|w)
            #p_d_on_w = P_wi_dj * P_wi
            pwi = P_wi_dj * pmi
            #print(f'pwi:{pwi} = P_wi_dj{P_wi_dj}* pmi {pmi}\n')
            
            words_in_t.append(pwi)
            #print(pwi)
            dj_wi_pwi_list.append(pwi)
    topic_pwi = sum(dj_wi_pwi_list)
    print(topic_pwi)
    total_pwi.append(topic_pwi)
```

```python
kpop_kr_score = sum(total_pwi)
print(kpop_kr_score)
```
