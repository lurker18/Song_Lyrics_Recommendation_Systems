```python
pattern = re.compile('([A-Z])')
only_english_numbers = re.compile("[^ a-zA-Z'\n0-9+]")
```

```python
re.sub(pattern, r" \1", df['Translations'][0])
```

```python
def cleanText(text):
    temp = re.sub(pattern, r" \1", text)
    cleaned = re.sub(only_english_numbers, '\n', temp)
    return cleaned
```

```python
df['Filtered'] = None
```

```python
for i in range(len(df)):
    df['Filtered'][i] = cleanText(df['Translations'][i])
```
