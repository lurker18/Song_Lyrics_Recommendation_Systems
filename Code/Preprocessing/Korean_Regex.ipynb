```python
only_korean_numbers = re.compile('[^ ㄱ-ㅣ가-힣,0-9+]')
```

```python
re.sub(only_korean_numbers, '', df['Translations'][1])
```

```python
def cleanText(text):
    temp = re.sub(only_korean_numbers, '', text)
    cleaned = temp.replace(',', '\n').replace('\n ','\n')
    return cleaned
```

```python
df['Filtered'] = None
```

```python
for i in range(len(df)):
    df['Filtered'][i] = cleanText(df['Translations'][i])
```
